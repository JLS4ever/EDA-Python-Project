**Segment Distance into Short, Medium, Long:**

```python
# Define bins
bins = [0, 1000, 3000, df['flight_distance'].max()]
labels = ['Short Haul', 'Medium Haul', 'Long Haul']

df['flight_category'] = pd.cut(df['flight_distance'], bins=bins, labels=labels)

print(df['flight_category'].value_counts())
```

**Compare Patterns Across Bins:**

```python
ct = pd.crosstab(df['flight_category'], df['satisfaction'], normalize='index') * 100
print(ct)

ct.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Satisfaction Rate by Flight Distance Category')
plt.ylabel('Percentage')
plt.legend(title='Satisfaction')
plt.show()
```

```python
ct_class = pd.crosstab(df['flight_category'], df['class'], normalize='index') * 100
print(ct_class)

ct_class.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Class Choice by Flight Distance Category')
plt.ylabel('Percentage')
plt.legend(title='Class')
plt.show()
```

```python
ct_travel = pd.crosstab(df['flight_category'], df['type_of_travel'], normalize='index') * 100
print(ct_travel)

ct_travel.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Type of Travel by Flight Distance Category')
plt.ylabel('Percentage')
plt.legend(title='Type of Travel')
plt.show()
```

**Our analysis shows that flight distance indeed affects passenger behavior and satisfaction. The distribution of flight distances is skewed, with many short-haul flights but a significant tail of longer routes. Passengers flying longer distances tend to choose higher travel classes like Business Class, suggesting comfort matters more on lengthy trips. Business travelers are more common on medium and long-haul flights, while personal travel dominates short-haul routes. Interestingly, satisfaction rates appear slightly higher among long-haul passengers, possibly reflecting better service or amenities offered on longer flights. By segmenting flights into short, medium, and long categories, we found clear differences in class preference, type of travel, and satisfaction, confirming that flight distance shapes how passengers travel and how they rate their experiences.**

**Which factors contribute to customer satisfaction the most? What about dissatisfaction?**

**Explore Correlations:**

```python
# Encode satisfaction into binary
df['satisfaction_binary'] = df['satisfaction'].map({
    'Neutral or Dissatisfied': 0,
    'Satisfied': 1
})

# Get numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Drop non-service columns if needed
exclude_cols = ['id', 'age', 'flight_distance', 'departure_delay', 'arrival_delay']
service_cols = [col for col in num_cols if col not in exclude_cols and col != 'satisfaction_binary']

corr = df[service_cols + ['satisfaction_binary']].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr[['satisfaction_binary']].sort_values(by='satisfaction_binary', ascending=False),
            annot=True, cmap='viridis')
plt.title('Correlation with Satisfaction')
plt.show()
```



*   Positive correlations → factors associated with satisfaction
*   Negative correlations → factors associated with dissatisfaction




**Compare Average Ratings for Satisfied vs Dissatisfied:**

```python
# Select only rating columns
ratings_cols = service_cols

group_means = df.groupby('satisfaction')[ratings_cols].mean().T
group_means.columns = ['Mean_Dissatisfied', 'Mean_Satisfied']
group_means['Diff'] = group_means['Mean_Satisfied'] - group_means['Mean_Dissatisfied']
group_means = group_means.sort_values(by='Diff', ascending=False)

print(group_means)
```

**Plot the Differences:**

```python
group_means['Diff'].plot(kind='barh', color='purple', figsize=(8,6))
plt.title('Difference in Average Ratings (Satisfied - Dissatisfied)')
plt.xlabel('Difference in Rating')
plt.show()
```

**Feature Importance via Random Forest:**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Prepare data
X = df[ratings_cols]
y = df['satisfaction_binary']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Get importances
importances = pd.Series(model.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=True)

# Plot
importances.plot(kind='barh', color='teal', figsize=(8,6))
plt.title('Feature Importance for Predicting Satisfaction')
plt.xlabel('Importance Score')
plt.show()
```

**Conclusions**

Our exploratory data analysis of airline passenger satisfaction revealed valuable insights into customer behavior and the factors shaping their experiences. Overall, about 55–60% of passengers report being satisfied, though satisfaction varies notably across different groups. Loyal customers, who are typically older, business travelers, and frequent flyers choosing longer routes and higher travel classes like Business Class, are far more satisfied than disloyal passengers. Flight distance emerged as a significant factor influencing travel patterns; longer flights are more associated with business travel, higher class choices, and slightly higher satisfaction levels. Delving into drivers of satisfaction, we found that service-related factors such as on-board service, seat comfort, cleanliness, and in-flight entertainment play the largest role in determining how passengers feel about their journeys. Conversely, issues like delays, poor food and drink, uncomfortable gate locations, and limited legroom contribute to dissatisfaction. These findings suggest that airlines aiming to boost customer satisfaction should prioritize service quality improvements, particularly for loyal and business travelers on longer flights, as well as maintain operational efficiency to reduce delays. Altogether, the analysis offers a roadmap for enhancing the passenger experience and fostering stronger customer loyalty.
